## Trying the Binary Classification Breast Cancer Dataset

```python
from sklearn.datasets import load_breast_cancer
```

```python
X, Y = load_breast_cancer(as_frame=True, return_X_y=True)
```

```python
X.head()
```

```python
X.describe()
```

```python
X.info()
```

```python
n_pos = (Y==1).sum()
n_neg = (Y==0).sum()
print(f"No. of positive samples in the dataset are {n_pos}\nNo. of negative samples in the dataset are {n_neg}")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state=42)
```

```python
for x in [X_train, X_test, y_train, y_test]:
    print(x.shape)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel = "linear", random_state=42, C = 1)
svc.fit(X_train, y_train)
score = svc.score(X_test, y_test)
score
```

```python
print(f"Therefore, we are able to achieve an accuracy of {np.round(score*100,1)} % on the test dataset")
```

```python
from sklearn.metrics import classification_report
print(classification_report(svc.predict(X_test), y_test))
```

## Taking an example to visualize the effects of gamma coefficient on the SVM classifier with rbf kernel function

```python
>>> X = np.c_[# negative class
... (.3, -.8),
... (-1.5, -1),
... (-1.3, -.8),
... (-1.1, -1.3),
... (-1.2, -.3),
... (-1.3, -.5),
... (-.6, 1.1),
... (-1.4, 2.2),
... (1, 1),
... # positive class
... (1.3, .8),
... (1.2, .5),
... (.2, -2),
... (.5, -2.4),
... (.2, -2.3),
... (0, -2.7),
... (1.3, 2.1)].T

Y = [-1]*8 + [1]*8
```

```python
def contour_plotter(model, title):
    x = np.linspace(-1.5, 1.5, 100)
    y = np.linspace(-3, 2.3, 100)
    x_mesh, y_mesh = np.meshgrid(x,y)
    x_mesh_flat, y_mesh_flat = x_mesh.ravel(), y_mesh.ravel()
    X_ = np.c_[x_mesh_flat, y_mesh_flat]
    z = model.decision_function(X_).reshape(x_mesh.shape)
    plt.contourf(x_mesh, y_mesh, z>0, cmap=plt.cm.Paired, zorder = 1)
    plt.scatter(X[:, 0], X[:, 1], c = ['b']*8 + ['r']*8, zorder = 2)
    plt.contour(x_mesh, y_mesh, z, colors = "black", levels = [-0.5, 0, 0.5], linestyles = ["--", "-", "--"], zorder = 3)
    plt.title(title)
    plt.show()
```

```python
gamma_values = [1,2,4]

for gamma in gamma_values:
    svc = SVC(kernel="rbf", gamma=gamma)
    svc.fit(X, Y)
    title = "Gamma = " + str(gamma)
    contour_plotter(svc, title)
```

## Using the SVM's for face recognition using LFW data

```python
## Downloading the Labeled Faces in the Wild Image dataset
from sklearn.datasets import fetch_lfw_people
lfw_dataset = fetch_lfw_people(min_faces_per_person=80) ##Downloading only persons with min 80 images
```

```python
lfw_images = lfw_dataset.images
```

```python
lfw_images.shape
```

```python
lfw_targets = lfw_dataset.target
lfw_targets
```

```python
# Creating a name dictionary to get people names from their target label
name_dict = dict(zip(np.unique(lfw_targets), lfw_dataset.target_names))
name_dict
```

```python
fig, ax = plt.subplots(2,6, figsize = (12,4))
for index, axis in enumerate(ax.ravel()):
    axis.imshow(lfw_images[index], cmap = "gray")
    axis.set(xticks=[], yticks = [], title = str(name_dict[lfw_targets[index]]))
```

```python
## Visualizing the number of pics for different labels
from collections import defaultdict

count_dict = defaultdict(int)
for label in np.unique(lfw_targets):
    count_dict[label] = (lfw_targets == label).sum()
plt.bar(list(name_dict.values()), list(count_dict.values()))
plt.xticks(rotation = 45)
plt.axhline(np.mean(list(count_dict.values())),color = "r", linestyle = "--") ## Average number of images for a class
plt.show()
```

From the above bargraph, we can see that the class distribution is not uniform. Label 2, i.e. GeorgeWBush, has the largest number of images, with more than double the average.

```python
X_train, X_test, y_train, y_test = train_test_split(lfw_images, lfw_targets, random_state = 42)

# Converting the three dimensional arrays containing images to two dimensional arrays
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)

for x in [X_train, X_test, y_train, y_test]:
    print(x.shape)
```

```python
## Using SVC for face recognition

from sklearn.svm import SVC
lfw_svc = SVC(class_weight="balanced", random_state = 42)
lfw_svc.fit(X_train, y_train)
accuracy = lfw_svc.score(X_test, y_test)
print(f"The model with default parameters is able to score an accuracy of {accuracy*100:.1f}%")
```

```python
## Carrying out hyperparameter tuning

from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10],'gamma': [1e-07, 1e-08, 1e-06], 'kernel' : ['rbf', 'linear'] }

grid_search = GridSearchCV(estimator = lfw_svc, param_grid=param_grid, cv = 5, verbose = 2)
grid_search.fit(X_train, y_train)
```

```python
## Best estimator parameters
print(grid_search.best_params_)

## Taking the best estimator as our model

lfw_svc = grid_search.best_estimator_

## Best parameter accuracy
print(f"Best model's accuracy on the test set is {lfw_svc.score(X_test, y_test)*100:.1f}%")
```

```python
## Printing the classification report for the best classifier

print(classification_report(lfw_svc.predict(X_test), y_test))
```

Applying PCA for better performance

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

# Reducing the no. of features from 2914 to 100
lfw_pca = PCA(n_components = 100, whiten = True, random_state = 42)
lfw_pipe = Pipeline([("pca", lfw_pca),
                   ("svc", SVC(kernel="rbf", class_weight="balanced", random_state=42))])
param_pipe = {'svc__C': [1, 3, 10], 'svc__gamma': [0.001, 0.005]}
grid_pipe = GridSearchCV(estimator = lfw_pipe, param_grid=param_pipe, cv = 5, verbose=2)
grid_pipe.fit(X_train, y_train)
```

```python
print(f"best parameters are: {grid_pipe.best_params_}")
lfw_svc_pipe = grid_pipe.best_estimator_
print(f"The SVC with PCA gets an accuracy of {lfw_svc_pipe.score(X_test, y_test)*100:.1f}% as opposed to the accuracy of {lfw_svc.score(X_test, y_test)*100:.1f} of the previous model without PCA")
```

```python
## Printing the classification report

print(classification_report(lfw_svc_pipe.predict(X_test), y_test))
```
