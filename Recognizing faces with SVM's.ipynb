## Trying the Binary Classification Breast Cancer Dataset

```python
from sklearn.datasets import load_breast_cancer
```

```python
X, Y = load_breast_cancer(as_frame=True, return_X_y=True)
```

```python
X.head()
```

```python
X.describe()
```

```python
X.info()
```

```python
n_pos = (Y==1).sum()
n_neg = (Y==0).sum()
print(f"No. of positive samples in the dataset are {n_pos}\nNo. of negative samples in the dataset are {n_neg}")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state=42)
```

```python
for x in [X_train, X_test, y_train, y_test]:
    print(x.shape)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel = "linear", random_state=42, C = 1)
svc.fit(X_train, y_train)
score = svc.score(X_test, y_test)
score
```

```python
print(f"Therefore, we are able to achieve an accuracy of {np.round(score*100,1)} % on the test dataset")
```

```python
from sklearn.metrics import classification_report
print(classification_report(svc.predict(X_test), y_test))
```

## Trying the Wine Dataset for Multiclass Classification

```python
from sklearn.datasets import load_wine
X, Y = load_wine(return_X_y=True, as_frame=True)
```

```python
X.head()
```

```python
X.describe()
```

```python
X.info()
```

```python
Y.unique()
```

```python
n_0 = (Y==0).sum()
n_1 = (Y==1).sum()
n_2 = (Y==2).sum()

print(f"The no. of different classes in the dataset are:\nClass_0: {n_0}, Class_1: {n_1}, Class_2: {n_2}\nTotal no. of entries = {len(Y)}")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state=42)
```

```python
for x in [X_train, X_test, y_train, y_test]:
    print(x.shape)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel = "linear", random_state=42, C = 1)
svc.fit(X_train, y_train)
score = svc.score(X_test, y_test)
print(f"Therefore, we are able to achieve an accuracy of {score*100:.1f} % on the test dataset")
```

```python
from sklearn.metrics import classification_report
print(classification_report(svc.predict(X_test), y_test))
```
